```python
df.columns = df.columns.str.strip()
```

```python
investor_columns = ['Anupam', 'Ashneer', 'Namita', 'Aman', 'Peyush', 'Vineeta', 'Ghazal']
for col in investor_columns:
    df[col] = df[col].apply(lambda x: 1 if x == 'Y' else 0)
```

```python
df['Total Investors'] = df[investor_columns].sum(axis=1)
```

```python
features = ['Equity', 'Debt (In lakhs INR)', 'Season', 'Total Investors']
target = 'Investment Amount (In Lakhs INR)'
```

```python
df['Equity'] = df['Equity'].str.rstrip('%').astype(float)
```

```python
X = df[features]
Y = df[target]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
